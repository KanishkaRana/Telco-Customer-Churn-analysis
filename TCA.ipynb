```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv( 'WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
df.info()
```

```python
# Replace all whitespace-only strings (including tabs, multiple spaces, etc.) with NaN
df["TotalCharges"] = df["TotalCharges"].replace(r"^\s*$", "0", regex=True)

# Convert column to float
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
df.isnull().sum()
```

```python
#descriptive analysis-
df.describe()
```

```python
df["customerID"].duplicated().sum() #NO DUPLICATE CUSTOMER ID
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
#converted 0and 1 to no and yes respectively to understand better
# Apply the function outside the function definition
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
#checking how many customers churned out
ax= sb.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#in percentage
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels=gb.index, autopct = "%1.2f%%")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sb.histplot(x= "tenure", data = df)
plt.show()
```
